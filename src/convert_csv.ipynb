# Utility to adapts csv
* source https://www.kaggle.com/datasets/nphantawee/pump-sensor-data

```python
import pandas as pd
import numpy as np
df = pd.read_csv("../data/sensors.csv")
df["timestamp"] = pd.to_datetime( df["timestamp"], format='mixed' ) + np.timedelta64(7 * 365,'D')
```

```python
df[["timestamp",	"sensor_00",	"sensor_01",	"sensor_02",
    "sensor_03",	"sensor_04",	"sensor_05",	"sensor_06",
    "sensor_07"]].to_csv("../data/ACME Limited.ACME Plant.PUMP-001.csv", index = False)
```
